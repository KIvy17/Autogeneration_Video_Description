```python
import os
import random
import cv2

# Paths to directories
train_folder = "/content/drive/MyDrive/rutube_hackathon_summary_generation_novosibirsk/train/train_video"
output_folder = "/content/drive/MyDrive/rutube_hackathon_summary_generation_novosibirsk/train/extracted_frames"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)
os.makedirs(os.path.join(output_folder, "folder1"), exist_ok=True)
os.makedirs(os.path.join(output_folder, "folder2"), exist_ok=True)
```

```python
# Function to extract frames from videos and save them as images
def extract_frames(video_file, output_folder, num_frames):
    cap = cv2.VideoCapture(video_file)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Randomly select frames
    selected_frames = random.sample(range(total_frames), min(num_frames, total_frames))

    # Extract and save selected frames as images
    for frame_idx in selected_frames:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
        ret, frame = cap.read()
        if ret:
            output_file = os.path.join(output_folder, f"{os.path.splitext(os.path.basename(video_file))[0]}_{frame_idx}.jpg")
            cv2.imwrite(output_file, frame)
```

```python
# Process each video file in the train folder
video_files = [os.path.join(train_folder, file) for file in os.listdir(train_folder) if file.endswith(".mp4")]

# Initialize counters for each folder
folder1_frames = 0
folder2_frames = 0
```

```python
# Process videos and distribute frames to folders
for idx, video_file in enumerate(video_files):
    if folder1_frames >= 500 and folder2_frames >= 500:
        break

    # Randomly determine the number of frames to extract (2 to 3 frames)
    num_frames = 2

    # Determine the folder to save frames based on the index
    save_to_folder = "folder1" if idx % 2 == 0 else "folder2"

    if save_to_folder == "folder1" and folder1_frames < 500:
        extract_frames(video_file, os.path.join(output_folder, "folder1"), num_frames)
        folder1_frames += num_frames
    elif save_to_folder == "folder2" and folder2_frames < 500:
        extract_frames(video_file, os.path.join(output_folder, "folder2"), num_frames)
        folder2_frames += num_frames

    # Print a message every 25th video
    if (idx + 1) % 25 == 0:
        print(f"Processed {idx + 1} videos and chosen frames.")
```

```python
import os
import cv2
import random

# ÐÐ°Ñ€ÐµÐ·Ð°ÐµÐ¼ Ð¿Ð¾ 2 ÐºÑƒÑÐºÐ° Ñ Ð²Ð¸Ð´ÐµÐ¾ Ñ€Ð°Ð·Ð¼ÐµÑ€Ð¾ Ð¿Ð¾ 30-40 ÑÐµÐº
def cut_video(video_path, output_folder, num_cuts=2, min_duration=30, max_duration=40):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps

    for i in range(num_cuts):
        # Randomly select start and end times for the cut
        start_time = random.uniform(0, max(0, duration - max_duration))
        end_time = min(start_time + random.uniform(min_duration, max_duration), duration)

        # Set the start frame
        cap.set(cv2.CAP_PROP_POS_MSEC, start_time * 1000)

        # Create output file path
        video_name = os.path.splitext(os.path.basename(video_path))[0]
        output_file = os.path.join(output_folder, f"{video_name}_cut{i + 1}.mp4")
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(output_file, fourcc, fps, (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
                                                         int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))

        # Read frames and write to output video until end time is reached
        while cap.get(cv2.CAP_PROP_POS_MSEC) <= end_time * 1000:
            ret, frame = cap.read()
            if not ret:
                break
            out.write(frame)

        # Release video capture and writer
        out.release()

    cap.release()

# Paths to videos
videos_folder = "/content/"
output_folder = "/content/drive/MyDrive/rutube_hackathon_summary_generation_novosibirsk/test/cut_videos"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# List of video fileÑ‹Ñ‹Ñ‹
videos = ["Elon_vid.mp4", "messi_vid.mp4"]


# Process each video
for video_file in videos:
    video_path = os.path.join(videos_folder, video_file)

    # Cut the video into two segments
    cut_video(video_path, output_folder)
```

```python
from google.colab.patches import cv2_imshow

# Function to apply object tracking to a video, draw bounding boxes, and save the annotated video
def apply_object_tracking(video_path, output_path):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    video_name = os.path.splitext(os.path.basename(video_path))[0]

    object_count = {}  # Dictionary to store object counts
    object_track = {}  # Dictionary to track object durations
    current_frame = 0   # Initialize current frame counter

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (int(cap.get(3)), int(cap.get(4))))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break


        results = model(frame)

        if results:
          # Access class names from the YOLO model
          class_names = model.names

          # Iterate over each result in the results list
          for result in results:
              # Access the boxes attribute for each result
              for box in result.boxes.data.tolist():
                  x1, y1, x2, y2, conf, cls = box
                  x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                  # Draw bounding boxes around detected objects
                  cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                  # Get the class name using cls as index
                  class_name = class_names[int(cls)]
                  cv2.putText(frame, f'{class_name} {conf:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

                  # Update object count and track
                  obj_id = f"{class_name}_{int(x1)}_{int(y1)}"  # Generate unique object ID
                  if obj_id not in object_count:
                      object_count[obj_id] = 1
                      object_track[obj_id] = {'start_frame': current_frame, 'end_frame': current_frame}
                  else:
                      object_count[obj_id] += 1
                      object_track[obj_id]['end_frame'] = current_frame

        # Write the frame to the output video
        out.write(frame)
        current_frame += 1

        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    # Output object counts and track durations
    for obj_id, count in object_count.items():
        print(f"Object {obj_id} appeared {count} times in the video.")
        track_duration = object_track[obj_id]['end_frame'] - object_track[obj_id]['start_frame']
        print(f"Object {obj_id} appeared for {track_duration} frames.")

    cap.release()
    out.release()
    cv2.destroyAllWindows()


video_path = '/content/drive/MyDrive/rutube_hackathon_summary_generation_novosibirsk/test/cut_videos/Elon_vid_cut2.mp4'
output_video_path = '/content/drive/MyDrive/rutube_hackathon_summary_generation_novosibirsk/test/cut_videos/anotated_Elon_vid_cut2.mp4'
apply_object_tracking(video_path, output_video_path)
```
